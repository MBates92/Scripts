```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as ss
from maths.fields.convolve import nd_convolve
import DeltaVar as dv
```

```python
H = np.array([0.0,0.0,0.0,0.5,0.5,0.5,1.0,1.0,1.0])
```

```python
X = []
for h in H:
    X.append(ss.fBm(1000,2,h,exp=False))
print(np.shape(X))
```

```python
L = 10**(np.linspace(np.log10(0.001),np.log10(1), 100))
v = 1.5
```

```python
deltavar = []
for j in range(len(H)):
    mh_dv = np.zeros(100)
    for i in range(100):
        print(j,i)
        mh_dv[i]=dv.DeltaVar(X[j],L[i],v=1.5,method='fourier')
    deltavar.append(mh_dv)
```

```python
plt.figure(figsize = (1080/144,1080/144),dpi=144)
[plt.plot(np.log10(L),np.log10(deltavar[i])) for i in range(len(H))]
plt.show()
```
