```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import SpectralSynthesis as ss
import FractalVariates as fv
```

```python
N_p = 100
E = 2
H = 0.5
sigma = 1.0
exp = True
```

```python
field = ss.fBm(N=N_p, E = E, H = H, exp = exp, sigma = sigma)

plt.figure(dpi=100)
plt.imshow(field, origin = 'lower')
plt.show()
```

```python
plt.figure(dpi=100)
plt.imshow(field)
plt.colorbar()
plt.show()
```

```python
star_positions = fv.variates(field, 300)

plt.figure(dpi=100)
plt.imshow(field, origin = 'lower')
plt.scatter(star_positions[:,0], star_positions[:,1], s = 0.5, c ='white')
plt.show()
```

```python
cv_data = np.zeros((N_p*2,N_p*2,3), np.uint8)

for position in star_positions:
    print(position)
    cv2.circle(cv_data, (int(position[0]*2), int(position[1]*2)), 1, (0,255,0), -1)
```

```python
flipped = cv2.flip(cv_data,0)
```

```python
resized = cv2.resize(flipped, dsize = None, fx=2, fy=2)

cv2.imshow('stars', resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('star_example.png', cv_data)
```
