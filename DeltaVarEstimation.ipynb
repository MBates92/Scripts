```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as SS
import DeltaVar as DV
```

```python
X1 = SS.fBm(1000,2,0.0,sigma= 2.0,exp = True)
X2 = SS.fBm(1000,2,0.5,sigma= 2.0,exp = True)
X3 = SS.fBm(1000,2,1.0,sigma= 2.0,exp = True)
```

```python
plt.imshow(X1, origin = 'lower')
plt.colorbar()
plt.show()
```

```python
plt.imshow(X2, origin = 'lower')
plt.colorbar()
plt.show()
```

```python
plt.imshow(X3, origin = 'lower')
plt.colorbar()
plt.show()
```

```python
L = 10**(np.linspace(np.log10(0.003),np.log10(0.1), 100))
v = 1.5
```

```python
sigma_d = []
for i in range(len(L)):
    sigma_d.append(DV.DeltaVar(X1,L[i],v, method = 'convolve'))
    print(i)
```

```python
sigma_d1 = sigma_d
```

```python
sigma_d2 = []
for i in range(len(L)):
    sigma_d2.append(DV.DeltaVar(X2,L[i],v, method = 'convolve'))
    print(i)
```

```python
sigma_d3 = []
for i in range(len(L)):
    sigma_d3.append(DV.DeltaVar(X3,L[i],v, method = 'convolve'))
    print(i)
```

```python
plt.plot(np.log10(L),np.log10(sigma_d1))
plt.plot(np.log10(L),np.log10(sigma_d2))
plt.plot(np.log10(L),np.log10(sigma_d3))
plt.grid()
plt.show()
```

```python
params1 = DV.HurstEstimator(sigma_d1,L)
params2 = DV.HurstEstimator(sigma_d2,L)
params3 = DV.HurstEstimator(sigma_d3,L)
```

```python
H_est1 = params1[0]
H_est2 = params2[0]
H_est3 = params3[0]
```

```python
print(H_est1)
print(H_est2)
print(H_est3)
```

```python
plt.figure(figsize=(1080/144, 1080/144), dpi=144)
plt.plot(np.log10(L),np.log10(sigma_d1), c = 'b')
plt.plot(np.log10(L),np.log10(sigma_d2), c = 'c')
plt.plot(np.log10(L),np.log10(sigma_d3), c = 'g')
plt.text(-2.5,12, r'$H_{est}= $'+'%.3f'%H_est1[0]+r'$\pm %.3f$'%H_est1[1], bbox=dict(facecolor='blue', alpha=0.3))
plt.text(-2.45,10, r'$H_{est}= $'+'%.3f'%H_est2[0]+r'$\pm %.3f$'%H_est2[1], bbox=dict(facecolor='cyan', alpha=0.3))
plt.text(-2.0,8.5, r'$H_{est}= $'+'%.3f'%H_est3[0]+r'$\pm %.3f$'%H_est3[1], bbox=dict(facecolor='green', alpha=0.3))
plt.title(r'Periodic, N=1000, $H = [0.0,0.5,1.0], \sigma = 2.0$')
plt.grid()
plt.ylabel(r'$\log(\sigma^2_\Delta(L))$')
plt.xlabel(r'$\log(L[pixels])$')
plt.savefig('../SpectralSynthesis/MiscImages/DeltaVarPeriodicSigmaLarge',bbox_inches="tight")
plt.show()
```
