```python
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import DeltaVar as dv

import SpectralSynthesis as ss
from tqdm import tqdm
```

```python
L = 10**(np.linspace(-1.9,-0.01, 300))
L_range = 10**(np.linspace(-1.5,-0.5,100))
v = 1.5
```

```python
H_actual = np.linspace(0,1,11)

print(H_actual)
```

```python
DATA=[]
for h in H_actual:
    field = ss.fBm(N=128, E=2, H=h, exp=False)
    m_1_field = np.mean(field)
    s_1_field = np.std(field)
    m_2_field = 0
    s_2_field = 1/4
        
    field *= s_2_field/s_1_field
    field += (m_2_field-m_1_field*s_2_field/s_1_field)
    DATA.append([field,h])

print(DATA)
```

```python
print(DATA[0])
```

```python
for img,h in DATA:
    plt.figure()
    plt.imshow(img)
    plt.title(h)
    plt.colorbar()
    plt.show()
```

```python
H_predicted = []
c=0
f,ax = plt.subplots(figsize=(1080/200,1080/200),dpi=200)
for img,H in DATA:
    img_array = img
    sigma_d = [dv.DeltaVar(img_array, l, v, periodicity = True) for l in L]
    sigma_d_range = [dv.DeltaVar(img_array, l, v, periodicity = True) for l in L_range]
    h = dv.HurstEstimator(sigma_d_range,L_range)
    H_predicted.append(h[0])
    ax.plot(np.log10(L),np.log10(sigma_d),label=r'${\beta}_{\rm EST}=$'+'{:.2f}'.format(h[0]*2+2)+', '+r'${\beta}_{\rm TRUE}=$'+'{:.2f}'.format(H*2+2))

ax.axvspan(np.log10(L_range[0]),np.log10(L_range[-1]), color = 'red', alpha = 0.2)
plt.grid()
plt.ylabel(r'$$')
plt.legend(fontsize ='small')
plt.xlabel(r'$\log_{10}(L)$')
plt.ylabel(r'$\log_{10}(\sigma^2_\Delta(L))$')
plt.tight_layout()
plt.savefig('D:/Bates_Lomax_Whitworth_Paper/DeltaVarInertialRange.png',bbox_inches ='tight')
plt.show()
```

```python
from Math import mean_square_error

val_loss = mean_square_error(H_actual,H_predicted)
print(val_loss)
```

```python
np.savetxt('D:/ExponentiatedNonPeriodic/predictions.txt',H_predicted)
np.savetxt('D:/ExponentiatedNonPeriodic/evaluations.txt',[val_loss])
```

```python
plt.figure(figsize = (1080/200,1080/200), dpi = 200)
plt.scatter(H_actual,H_predicted)
plt.text(0.01, 0.95, 'RMSE = '+ '{:.3g}'.format(np.sqrt(val_loss)))
plt.plot([0,1],[0,1],c='k')
plt.ylim([0,1])
plt.xlim([0,1])
plt.xlabel(r'$H_{actual}$')
plt.ylabel(r'$H_{predicted}$')
plt.show()
```
