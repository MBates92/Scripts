```python
import numpy as np
import matplotlib
#matplotlib.use('agg')
import matplotlib.pyplot as plt
#matplotlib.rcParams.update({'font.size': 20})
import SpectralSynthesis as SS
```

```python
Seed = 10
N = 1000
```

```python
H1 = SS.fBm(N=N,E=1,H=0.0,Seed=Seed,exp=False)
H2 = SS.fBm(N=N,E=1,H=0.5,Seed=Seed,exp=False)
H3 = SS.fBm(N=N,E=1,H=1.0,Seed=Seed,exp=False)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3,1,figsize=(1280/144, 720/144), dpi=144)
ax1.plot(H1.real, 'k')
ax1.text(-40, -3.0, r'$\beta = {}$'.format(int(0*2+1)),bbox = dict(facecolor='white'))
ax1.set_ylabel('X(t)')
ax1.set_xticks([])
ax2.plot(H2.real, 'k')
ax2.text(-40, -2.0, r'$\beta = {}$'.format(int(0.5*2+1)),bbox = dict(facecolor='white'))
ax2.set_ylabel('X(t)')
ax2.set_xticks([])
ax3.plot(H3.real,'k')
ax3.text(-40, -1.6, r'$\beta = {}$'.format(int(1*2+1)),bbox = dict(facecolor='white'))
ax3.set_ylabel('X(t)')
ax3.set_xlabel('t')

f.subplots_adjust(wspace=0, hspace=0)

plt.savefig('D:/Bates_Lomax_Whitworth_Paper/1DfBm.png',bbox_inches="tight")
plt.show()
```
