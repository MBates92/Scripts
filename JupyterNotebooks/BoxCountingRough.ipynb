```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis1D as ss
```

```python
H=0.5
N=100
```

```python
X = ss.fbm(N,H).real
x = np.linspace(-1,1,N)
```

```python
X /= np.max(np.abs(X))
```

```python
plt.figure(figsize = (1920/144,1920/144), dpi=144)
ax = plt.axes()
plt.plot(x,X, linewidth = 1.0)
ax.axis('tight')
ax.axis('off')
plt.axis('equal')
plt.savefig('out.png', bbox_inches="tight")
plt.show()
```

```python
curve = plt.imread('out.png')
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
curve = rgb2gray(curve)*255
```

```python
print(np.shape(curve))
```

```python
plt.figure(figsize = (1920/144,1920/144), dpi=144)
plt.imshow(curve,cmap = plt.get_cmap('gray'))
plt.colorbar()
```

```python
def boxcount(Z, k):
    S = np.add.reduceat(
        np.add.reduceat(Z, np.arange(0, Z.shape[0], k), axis=0),
                           np.arange(0, Z.shape[1], k), axis=1)
    return len(np.where((S > 0) & (S < k*k))[0])
```

```python
curve = curve<0.9
```

```python
plt.figure(figsize = (1920/144,1920/144), dpi=144)
plt.imshow(curve)
plt.colorbar()
```
