```python
import numpy as np
import matplotlib.pyplot as plt
import DeltaVar as DV
```

```python
A = np.zeros((1000,1000))
A[500,500] = 255
```

```python
plt.imshow(A)
plt.xlim(450,550)
plt.ylim(450,550)
```

```python
shape = np.shape(A)
N = shape[0]
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)
xv, yv = np.meshgrid(x,y)

xv = xv - np.max(xv)/2
yv = yv - np.max(yv)/2

r = np.sqrt(xv**2 + yv**2)
r /= np.max(r)
```

```python
plt.imshow(r)
plt.colorbar()
```

```python
L = 0.1
v = 1.5
```

```python
kernel = DV.MexicanHat(r,L,v)
```

```python
plt.imshow(kernel)
plt.colorbar()
```

```python
from astropy.convolution import convolve_fft as convolve
```

```python
new_kernel = convolve(A,kernel,normalize_kernel=False, nan_treatment = 'fill')
```

```python
plt.imshow(new_kernel)
plt.colorbar()
```

```python
plt.imshow(new_kernel/np.max(new_kernel)-kernel/np.max(kernel), origin = 'lower')
plt.colorbar()
```
