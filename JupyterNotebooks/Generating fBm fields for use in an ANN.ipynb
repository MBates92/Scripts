```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as ss
```

```python
H_min = 0.0
H_max = 1.0

sigma_min = 0.0
sigma_max = 3.0
```

```python
N = 20000

H_list = np.random.random(N)
sigma_list = (sigma_max-sigma_min)*np.random.random(N)+sigma_min

plt.figure()
plt.hist(H_list,bins = 10)

plt.figure()
plt.hist(sigma_list, bins=10)
```

```python
target_list = []
target_list.append(H_list)
target_list.append(sigma_list)
print(target_list)
```

```python
plt.figure()
plt.hist(target_list[0],bins = 10)

plt.figure()
plt.hist(target_list[1], bins=10)
```

```python
N_p = 100
E = 2
```

```python
path = '../Data/fBmFields/2DExponentiated/100px/Training/features/'
for i in range(N):
    field = ss.fBm(N_p,E,H=target_list[0][i],sigma=target_list[1][i])
    field = field-np.amin(field)
    field /= np.amax(field)
    np.savetxt('{}field_{num:06d}.txt'.format(path,num=i),field)
```

```python
path = '../Data/fBmFields/2DExponentiated/100px/Training/labels/'
np.savetxt('{}labels.txt'.format(path),np.asarray(target_list).T)
```
