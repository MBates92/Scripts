```python
import numpy as np
import matplotlib.pyplot as plt
import DeltaVar as dv
```

```python
N = 1000
x = np.linspace(-2**-0.5,2**-0.5,N, dtype=np.float64)
y = np.linspace(-2**-0.5,2**-0.5,N, dtype=np.float64)
xv, yv = np.meshgrid(x,y)

r = np.sqrt(xv**2 + yv**2)
r /=np.max(r)
```

```python
f = dv.MexicanHat(r,0.5)

fig = plt.figure(frameon=False)
fig.set_size_inches(10,10)
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(f, aspect='equal',cmap='hot')
fig.savefig('MHat.png', dpi=200)
```

```python
f1D = f[:,499]
```

```python
plt.figure(dpi=200)
plt.plot(f1D)
plt.savefig('MHat1D.png')
plt.show()
```

```python
import SpectralSynthesis as ss

X = ss.fBm(N=1000, E=2, exp = False, H = 0.5)
```

```python
f_fft = np.fft.fftn(f)
X_fft = np.fft.fftn(X)
```

```python
prod = f_fft*X_fft
conv = np.fft.ifftn(prod).real
```
