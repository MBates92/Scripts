```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as ss
import DeltaVar as dv
```

```python
H = np.linspace(0,1,11)
N = 5
L = 10**(np.linspace(np.log10(0.0005),np.log10(1), 100))
v = 1.5
N_grid = 2000
```

```python
X = []
for h in H:
    X_i = []
    for i in range(N):
        X_i.append(ss.fBm(N_grid,2,h,exp=False))
        print(h,i)
    X_i = np.asarray(X_i)
    X.append(X_i)
print(np.shape(X))
```

```python
plt.imshow(X[10][0,:,:], origin = 'lower')
```

```python
for x in X:
    for i in range(5):
        x[i,:,:] = np.roll(x[i,:,:],int(N_grid/2),axis = (0,1))
        x[i,:,int(N_grid/2):] = 0
        x[i,int(N_grid/2):,:] = 0
```

```python
plt.imshow(X[10][0,:,:], origin = 'lower')
```

```python
deltavar = []
for i in range(len(H)):
    deltavar_i = []
    for j in range(N):
        deltavar_ij = []
        for l in L:
            deltavar_ij.append(dv.DeltaVar(X[i][j,:,:],l,v,method = 'fourier'))
            print(i,j,l)
        deltavar_ij = np.asarray(deltavar_ij)
        deltavar_i.append(deltavar_ij)
    deltavar_i = np.asarray(deltavar_i)
    deltavar.append(deltavar_i)
print(np.shape(deltavar))
```

```python
print(H)
```

```python
[np.savetxt('../Data/H_estVSsigma/deltavar_nonperiodic_s'+'{0:02d}'.format(int(H[i]*10))+'.txt',deltavar[i],header = 'Delta variance for 5 non-periodic fBm fields with H = '+str(H[i])) for i in range(len(H))]
```

```python
plt.figure(figsize=(1080/144,1080/144),dpi=144)
colors = ['b', 'orange', 'r', 'green', 'cyan', 'gray', 'navy', 'magenta', 'yellow', 'olive', 'lime','purple']
for i in range(11):
    plt.figure(figsize=(1080/144,1080/144),dpi=144)
    for j in range(5):
        plt.plot(np.log10(L),np.log10(deltavar[i][j,:]), label = 'H = ' +str(H[i]), c = colors[i])
    plt.xlabel(r'$\log_{10}(L)$')
    plt.ylabel(r'$\log_{10}(\sigma^2_\Delta(L))$')
    plt.legend(loc = 4)
    plt.grid()
    frame1 = plt.gca()
    frame1.set_aspect('equal', adjustable='box')
    plt.savefig('../Images/DeltaVarRawNonPeriodic'+'{0:02d}'.format(int(H[i]*10))+'.png')
    plt.show()
```
