```python
import numpy as np
import matplotlib.pyplot as plt
import Features as features
from scipy.spatial import ConvexHull
```

```python
x = np.random.random(5)
y = np.random.random(5)
```

```python
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
points = np.array([x,y]).T
hull = ConvexHull(points)
points/=np.sqrt(hull.volume)
print(points)
```

```python
complete_x,complete_y = features.completeGraph(points,normalise = True)
```

```python
plt.figure()
plt.plot(complete_x,complete_y, c='r',lw = 1,alpha=0.5)
plt.scatter(points[:,0],points[:,1],c='k',s=2) 
plt.axis('equal')
plt.show()
```

```python
del_x,del_y = features.DelaunayTriangulation(points)
```
