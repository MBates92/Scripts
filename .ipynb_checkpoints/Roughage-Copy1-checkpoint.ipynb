```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import Math
import SpectralSynthesis1D as ss
```

```python
N =10000
```

```python
pn = ss.pink(N)
wn = ss.white(N)
bn = ss.fbm(N,0.5)
```

```python
x = np.linspace(10**-10,10**0,1000)
iwn = 1/(x**0)
ipn = 1/(x**1)
ibn = 1/(x**2)
```

```python
from scipy import signal
```

```python
fs = 1.0
fwn, Pxx_den_wn = signal.periodogram(wn, fs)
```

```python
fpn, Pxx_den_pn = signal.periodogram(pn, fs)
```

```python
fbn, Pxx_den_bn = signal.periodogram(bn.real, fs)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3,1, sharey = True, sharex = True,figsize=(1920/144, 1080/144), dpi=144)
ax1.grid()
ax3.grid()
ax2.grid()
ax1.loglog(fwn,Pxx_den_wn, label = r'$S(f)$')
ax1.loglog(x,iwn,label = r'$1/f^0$')
ax1.set_title('White Noise Spectrum')
ax1.set_ylabel('S(f)')
ax1.legend()
ax2.loglog(fpn.real, Pxx_den_pn.real, label = r'$S(f)$')
ax2.loglog(x,ipn/10**1,label = r'$1/f^1$')
ax2.set_title('Pink Noise Spectrum')
ax2.set_ylabel('S(f)')
ax2.legend()
ax3.loglog(fbn.real, Pxx_den_bn.real, label = r'$S(f)$')
ax3.loglog(x,ibn/10**13,label = r'$1/f^2$')
ax3.set_title('Brown Noise Spectrum')
ax3.set_ylabel('S(f)')
ax3.set_xlabel('f')
ax3.legend(loc = 3)

plt.ylim(10**-20,10**5)
plt.xlim(10**-3,max(fbn.real))
plt.savefig('../Thesis/Images/NoiseSpectrum.png')
plt.show()
```
