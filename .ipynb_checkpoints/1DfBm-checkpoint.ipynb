```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as SS
```

```python
Seed = 500
N = 1000
```

```python
H1 = SS.fbm(N,0.0,Seed)
H2 = SS.fbm(N,0.5,Seed)
H3 = SS.fbm(N,1.0,Seed)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3,1, sharex = True,figsize=(1920/144, 1080/144), dpi=144)
ax1.grid()
ax2.grid()
ax3.grid()
ax1.plot(H1.real, 'k')
ax1.set_title(r'$H=0.0$')
ax1.set_ylabel('X(t)')
ax2.plot(H2.real, 'k')
ax2.set_title(r'$H=0.5$')
ax2.set_ylabel('X(t)')
ax3.plot(H3.real,'k')
ax3.set_title(r'$H=1.0$')
ax3.set_ylabel('X(t)')

#plt.savefig('../Thesis/Images/1DfBm.png')
plt.show()
```
