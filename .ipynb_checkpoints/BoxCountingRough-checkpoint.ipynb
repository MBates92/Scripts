```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis1D as ss
```

```python
H=0.5
N=100
```

```python
X = ss.fbm(N,H).real
x = np.linspace(-1,1,N)
```

```python
X /= np.max(np.abs(X))
```

```python
plt.figure(figsize = (1080/144,1080/144), dpi=144)
ax = plt.axes()
plt.plot(x,X, linewidth = 0.5)
ax.axis('tight')
ax.axis('off')
plt.axis('equal')
plt.savefig('out.png', bbox_inches="tight")
plt.show()
```
