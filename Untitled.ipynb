```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltavar = np.loadtxt('../Data/H_estVSsigma/deltavar_nonperiodic_s05.txt')
```

```python
print(np.shape(deltavar))
```

```python
L = np.loadtxt('../Data/H_estVSsigma/L.txt')
print(np.shape(L))
```

```python
plt.figure(dpi=144)
plt.plot(np.log10(L),np.log10(deltavar[0]))
plt.grid()
plt.show()
```

```python
from scipy.interpolate import UnivariateSpline
```

```python
spline = UnivariateSpline(np.log10(L),np.log10(deltavar[0]),s=0,k=4)
```

```python
plt.figure(dpi=144)
plt.plot(np.log10(L),np.log10(deltavar[0]))
plt.plot(np.log10(L),spline(np.log10(L)))
plt.grid()
plt.show()
```

```python
spline_1o_deriv = spline.derivative(n=1)
spline_2o_deriv = spline.derivative(n=2)
spline_3o_deriv = spline.derivative(n=3)
print(spline_1o_deriv)
```

```python
plt.figure(dpi=144)
plt.plot(np.log10(L),spline(np.log10(L)))
plt.grid()
plt.show()

plt.figure(dpi=144)
plt.plot(np.log10(L),spline_2o_deriv(np.log10(L)))
plt.grid()
plt.show()
```
