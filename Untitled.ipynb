```python
import numpy as np
import matplotlib.pyplot as plt
import SpectralSynthesis as ss
```

```python
sigma = np.loadtxt('../Data/H_estVsSigma/sigma.txt')
H_est_s1 = np.loadtxt('../Data/H_estVsSigma/H_est_s1.txt')
H_est_s2 = np.loadtxt('../Data/H_estVsSigma/H_est_s2.txt')
H_est_s3 = np.loadtxt('../Data/H_estVsSigma/H_est_s3.txt')
H_est1 = np.loadtxt('../Data/H_estVsSigma/H_est1.txt')
H_est2 = np.loadtxt('../Data/H_estVsSigma/H_est2.txt')
H_est3 = np.loadtxt('../Data/H_estVsSigma/H_est3.txt')
```

```python
print(np.shape(H_est1))
```

```python
plt.figure(figsize=(1080/144, 1080/144), dpi=144)
ax = plt.axes()
plt.errorbar(sigma,H_est1[:,0],yerr= H_est1[:,1], c = 'blue')
plt.errorbar(sigma,H_est2[:,0],yerr= H_est2[:,1], c = 'orange')
plt.errorbar(sigma,H_est3[:,0],yerr= H_est3[:,1], c = 'green')
ax.axhline(y=H_est_s1[0],c = 'blue',linewidth = 0.7)
ax.axhspan(ymin=H_est_s1[0]-H_est_s1[1], ymax=H_est_s1[0]+H_est_s1[1], color='blue', alpha = 0.5)
ax.axhline(y=H_est_s2[0],c = 'orange',linewidth = 0.7)
ax.axhspan(ymin=H_est_s2[0]-H_est_s2[1], ymax=H_est_s2[0]+H_est_s2[1], color='orange', alpha = 0.5)
ax.axhline(y=H_est_s3[0],c = 'green',linewidth = 0.7)
ax.axhspan(ymin=H_est_s3[0]-H_est_s3[1], ymax=H_est_s3[0]+H_est_s3[1], color='green', alpha = 0.5)
plt.ylabel(r'$H_est$')
plt.xlabel(r'$\sigma$')
plt.text(0.054,0.05, r'$H=1.0$', bbox=dict(facecolor='green', alpha=0.3),transform = ax.transAxes)
plt.text(0.17,0.05, r'$H=0.5$', bbox=dict(facecolor='orange', alpha=0.3),transform = ax.transAxes)
plt.text(0.286,0.05, r'$H=0.0$', bbox=dict(facecolor='blue', alpha=0.3),transform = ax.transAxes)
plt.grid()
plt.show()
```
