```python
import numpy as np
import matplotlib.pyplot as plt
import Features as features
from scipy.spatial import ConvexHull
```

```python
x = np.random.random(7)
y = np.random.random(7)
```

```python
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
points = np.array([x,y]).T
hull = ConvexHull(points)
points/=np.sqrt(hull.volume)
print(points)
```

```python
complete_x,complete_y = features.completeGraph(points,normalise = True)
```

```python
del_x,del_y = features.DelaunayTriangulation(points)
```

```python
mst_x,mst_y = features.MinSpanTree(points,normalise=True)
```

```python
plt.figure(dpi = 144)
plt.plot(complete_x,complete_y, c='r',lw = 1,zorder=0)
plt.scatter(points[:,0],points[:,1],c='k',s=4,zorder=1)
frame1 = plt.gca()
frame1.set_aspect('equal', adjustable='box')
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)
plt.tight_layout()
plt.savefig('Images/CompleteExample.png')
plt.show()
```

```python
plt.figure(dpi = 144)
plt.plot(del_x,del_y, c='r',lw = 1,zorder=0)
plt.scatter(points[:,0],points[:,1],c='k',s=4,zorder=1)
frame1 = plt.gca()
frame1.set_aspect('equal', adjustable='box')
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)
plt.tight_layout()
plt.savefig('Images/DelaunayExample.png')
plt.show()
```

```python
plt.figure(dpi = 144)
plt.plot(mst_x,mst_y, c='r',lw = 1,zorder=0)
plt.scatter(points[:,0],points[:,1],c='k',s=4,zorder=1)
frame1 = plt.gca()
frame1.set_aspect('equal', adjustable='box')
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)
plt.tight_layout()
plt.savefig('Images/MSTExample.png')
plt.show()
```
